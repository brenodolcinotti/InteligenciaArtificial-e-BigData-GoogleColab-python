```python
# Importando as bibliotecas
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# Base de dados
dados = {
'idade': [25, 30, 45, 50, 60, 35, 40, 55],
'imc': [22, 28, 32, 35, 30, 26, 33, 29],
'historico_familiar': [0, 0, 1, 1, 1, 0, 1, 1],
'diabetes': [0, 0, 1, 1, 1, 0, 1, 1] # 1 = sim, 0 = não
}

df = pd.DataFrame(dados)

# Separar X e y
X = df[['idade', 'imc', 'historico_familiar']]
y = df['diabetes']

# Criar e treinar a árvore
arvore = DecisionTreeClassifier(random_state=42)
arvore.fit(X, y)

# Novo usuário
novo_usuario = pd.DataFrame({'idade': [42], 'imc': [31], 'historico_familiar': 1})
previsao_codificada = arvore.predict(novo_usuario)
previsao_final = 'Sim' if previsao_codificada[0] == 1 else 'Não'

print("Previsão Final - Este usuário tem diabetes?", previsao_final)
```
