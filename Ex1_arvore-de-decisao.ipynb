```python
# Importando as bibliotecas
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# Base de dados
dados = {
'idade': [25, 45, 35, 50, 23, 40, 60, 30],
'renda': [2000, 8000, 5000, 10000, 1800, 6000, 12000, 3000],
'historico_bom': [1, 1, 0, 1, 0, 1, 1, 0],
'aprovado': [0, 1, 0, 1, 0, 1, 1, 0] # 1 = aprovado, 0 = negado
}

df = pd.DataFrame(dados)

# Separar X e y
X = df[['idade', 'renda', 'historico_bom']]
y = df['aprovado']

# Criar e treinar a árvore
arvore = DecisionTreeClassifier(random_state=42)
arvore.fit(X, y)

# Novo usuário
novo_usuario = pd.DataFrame({'idade': [35], 'renda': [4500], 'historico_bom': [0]})
previsao_codificada = arvore.predict(novo_usuario)
previsao_final = 'Sim' if previsao_codificada[0] == 1 else 'Não'

print("Previsão Final - Este usuário está aprovado?", previsao_final)
```
