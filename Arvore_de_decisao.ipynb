```python
# Importando as bibliotecas
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# Base de dados fictícia
dados = {
    'idade': [15, 18, 25, 30, 40, 50, 60],
    'gosta_acao': [1, 1, 1, 0, 0, 0, 0], # 1 = gosta e 0 = não gosta
    'historico_credito': [1, 1, 1, 1, 0, 0, 0] # 1 = Bom e 0 = Ruim
}

df = pd.DataFrame(dados)

# Separar X e y
X = df[['idade', 'gosta_acao']]
y = df['vai_cinema']

# Criar e treinar a árvore
arvore = DecisionTreeClassifier(random_state=42)
arvore.fit(X, y)

# Novo usuário
novo_usuario = pd.DataFrame({'idade': [37], 'gosta_acao': [1]})
previsao_codificada = arvore.predict(novo_usuario)
previsao_final = 'Sim' if previsao_codificada[0] == 1 else 'Não'

print("Previsão Final - Este usuário gosta de ir ao cinema?", previsao_final)
```
